```python
#Перенесем 20% в папку test
import os
import random
import shutil

dataset_path = "/content/dataset"

# Папки с данными
train_folder = os.path.join(dataset_path, "train")
test_folder = os.path.join(dataset_path, "test")
```

```python
for class_folder in os.listdir(train_folder):
    class_path = os.path.join(train_folder, class_folder)
    if os.path.isdir(class_path):
        # Получаем список файлов в текущей папке класса
        files = os.listdir(class_path)
        # в test (20%)
        num_files_to_move = int(0.2 * len(files))
        # Выбираем случайные файлы
        files_to_move = random.sample(files, num_files_to_move)
        # Перемещаем выбранные файлы в папку test
        for file_name in files_to_move:
            src = os.path.join(class_path, file_name)
            dst = os.path.join(test_folder, class_folder, file_name)
            # Создадим папку класса в папке test
            if not os.path.exists(os.path.join(test_folder, class_folder)):
                os.makedirs(os.path.join(test_folder, class_folder))
            shutil.move(src, dst)
            print(f"Пемерещено фото: {src} -> {dst}")

print("Готово!")
```

Инициация скаченной модели

```python
model_path = '/content/YOLO8_my_cls.pt'
model = YOLO(model_path)
```

```python
import os
import pandas as pd
import numpy as np
import torch
```

```python
test_images_path1 = '/content/test_stg1'
test_images_path2 = '/content/test_stg2'

# Список всех файлов изображений
image_files1 = sorted(os.listdir(test_images_path1))
image_files2 = sorted(os.listdir(test_images_path2))

# Чтение существующего CSV файла
results_df = pd.read_csv('/content/sample_submission_stg2.csv')
```

```python
def process_images(image_files, images_path, results_df):
    for image_file in image_files:
        image_path = os.path.join(images_path, image_file)
        results = model(image_path)

        # Получение предсказаний модели
        preds = results[0].probs.data.cpu().numpy()

        # Обновление значений в DataFrame
        results_df.loc[results_df['image'] == image_file, ['ALB', 'BET', 'DOL', 'LAG', 'NoF', 'OTHER', 'SHARK', 'YFT']] = preds

    return results_df
```

```python
results_df = process_images(image_files1, test_images_path1, results_df)
results_df = process_images(image_files2, test_images_path2, results_df)
```

```python
output_csv_path = '/content/sample_submission_stg_output.csv'
results_df.to_csv(output_csv_path, index=False)
```
